### 2.3 - Linear Regression - Example 

### 2.3.1 - Question


Describe association between adult **height** and **weight**

### 2.3.2 - Scientific Model

- How does **height** influce **weight**?
- H --> W
- W = f(H)
- Weight is some function of height

- H --> W <-- U (Unobserved)
- W = f(H,U)
- Weight is some function of height and unobserved stuff

### 2.3.3 - Generative Model: H --> W

- H --> W <-- U (Unobserved)
- For adults, weight is a proportion of height plus the influence of unobserved causes
$$
W = \beta H + U
$$

### ************* GENERATIVE CODE *************

```r
##- Function to simulate weights of individuals from heights
# variables: height(H), weight(W), standard deviation (sd), beta

weight_simulation <- function(H,b,sd){
    # Generate random noise with mean 0 and standard deviation 'sd'
    U <- rnorm(length(H), mean=0, sd=sd)
  
    # Calculate weight as a linear function of height (H) with noise (U)
    W <- b * H + U
    return(W)
}
```

**Function Parameters:**

- **H:** A vector of height values.
- **b:** The beta coefficient.
- **sd:** Standard deviation used in generating the noise.

**rnorm Usage:**

- `rnorm(length(H), mean = 0, sd = sd)` generates random normal values for each height in H.

**Weight Calculation:**

- The weight `W` is computed as `b * H` plus the random noise `U`.


```r
H <- runif(200, min=130, max=170)
H
```

**runif(200, min = 130, max = 170)** generates 200 random numbers that are uniformly distributed between 130 and 170.

W <- weight_simulation(H, b=0.5, sd=5)
W

```r
plot(W ~ H, col=2, lwd=4)
```

In R, the **tilde (~) is used to create a formula**. In the context of plot(W ~ H), the formula indicates that W (weight) is being plotted as a function of H (height). This is a concise way to specify the relationship between the two variables, where W is treated as the dependent variable and H as the independent variable.

## 2.3.4 - Describing models

**Reasoned about formula usage in R for 4 seconds**

This diagram describes a simple data-generation process where:

- $H_i$ is a height value drawn from a uniform distribution on the interval \([130, 170]\).
- $\beta$ is a coefficient indicating how weight depends on height.
- $U_i$ is random noise, assumed to follow a normal distribution with mean 0 and standard deviation \(\sigma\).

Putting these together, each weight value \(W_i\) is modeled as a linear function of height \((\beta H_i)\) plus the random noise term \(U_i\).

The tilde $\sim$ in the expressions

$$
U_i \sim \text{Normal}(0, \sigma)
$$

$$
H_i \sim \text{Uniform}(130, 170)
$$

indicates that $U_i$ is distributed as a normal random variable with mean 0 and standard deviation ($\sigma$), and $H_i$ is distributed as a uniform random variable between 130 and 170.
